```python
# importeren benodigde packages
# import necessary packages
import pandas as pd
import numpy as np
```

```python
# path inladen
# load path
annotate_path = "~/Projecten/programma-discriminatie-en-racisme/datasets/Inter_annotator_ingevuld/"

# documenten laden
# load documents
annotated_tess = pd.read_excel(annotate_path + 'inter_annotate_Tess.xlsx')
annotated_milena = pd.read_excel(annotate_path + 'inter_annotate_Milena.xlsx')
annotated_michelle = pd.read_excel(annotate_path + 'inter_annotate_Michelle.xlsx')

# mergen data
# merge data
def merge_annotated_data(naam_dataset, n_annotators, n_annotations = 3):
    """ functie die de gesplitste data gesplitst door split_annotated_data weer kan samenvoegen.
    De datasets met de naam naam_dataset_n_annotators.csv worden ingeladen. Vervolgens wordt de oorspronkelijke dataset gereconstrueerd
    uit de ingeladen data. De oorspronkelijke zinnen en het overzicht van de labels wordt gereturned.

    Input:
    naam_dataset: str, de naam van de documenen die de annotators hebben ingevuld
    n_annotators: int, het aantal annotators
    n_annotations: int, het aantal annotations per zin. default 3

    Output:
    merged_data: een DataFrame met daarin de oorspronkelijke data en de n_annotations labels die zijn toegekend.
    """

    """ function that can merge data split by split_annotated_data.
    The datasets with the name naam_dataset_n_annotators.csv are loaded. The original dataset then gets reconstructed
    from the loaded data. The original sentences and the overview of the original labels is returned.
    """
    
    # inladen path naar data en creëren lege frames
    # load path to data and create empty frames
    dataset_path = "~/Projecten/programma-discriminatie-en-racisme/datasets/Datasets_1/"
    merged_data = pd.DataFrame()

    for i in range(n_annotators):
        # inladen annotated documenten en samenvoegen
        # load annotated documents and compile
        annotated_doc = pd.read_excel(dataset_path + naam_dataset + "_" + str(i) + '.xlsx')
        annotated_doc = annotated_doc.dropna(subset=['Label', 'text', 'Column1'])
        columns = list(annotated_doc.columns)

        #display(annotated_doc)

        # document opstellen
        # create document
        if i == 0:
            merged_data = annotated_doc
        else:
            merged_data = pd.concat([merged_data, annotated_doc])

    display(merged_data)

    #merged_data = merged_data.groupby(['Column1']).agg({'id': 'last', 'titel': 'last', 'text': 'last', 'zin': 'last', 'Label': 'mean'})
    #merged_data = merged_data.sort_values(by = ['Column1'])

    return pd.DataFrame(merged_data)

merged = merge_annotated_data('te_labellen', n_annotators = 8, n_annotations = 3)

display(merged)

# vind welke data gelabeld is en door wie
# find what data has been annotated and by whom
inter_annotate_1 = merged[merged.labeller_1 == 2.0][0:50] # Tess
inter_annotate_2 = merged[merged.labeller_1 == 3.0][0:25] # Cheyenne
inter_annotate_3 = merged[merged.labeller_1 == 4.0][0:50] # Michelle
inter_annotate_4 = merged[merged.labeller_1 == 5.0][0:25] # Ailing
inter_annotate_5 = merged[merged.labeller_1 == 6.0][0:25] # Amy
inter_annotate_6 = merged[merged.labeller_1 == 7.0][0:25] # Kiran
inter_annotate_7 = merged[merged.labeller_1 == 8.0][0:25] # Milena
```

```python
# samenvoegen alle gelabelde data
# compile all annotated data
annotated = pd.concat([annotated_tess, annotated_michelle, annotated_milena, inter_annotate_1, inter_annotate_2, inter_annotate_3, inter_annotate_4, inter_annotate_5, inter_annotate_6, inter_annotate_7])
annotated['Column1'] = pd.to_numeric(annotated['Column1'])
annotated = annotated.sort_values(by = ['Column1'])

label1 = []
label2 = []
label3 = []

# samenvoegen alle labels van annotators
# compile all labels from different annotators
for i in range(0, len(annotated), 3):
    label1.append(annotated['Label'].iloc[i])
    
for i in range(1, len(annotated), 3):
    label2.append(annotated['Label'].iloc[i])
    
for i in range(2, len(annotated), 3):
    label3.append(annotated['Label'].iloc[i])
    
annotated = annotated.groupby(['Column1']).agg({'id': 'last', 'titel': 'last', 'text': 'last', 'zin': 'last', 'Label': 'mean'})

annotated['Label1'] = label1
annotated['Label2'] = label2
annotated['Label3'] = label3

display(annotated)

# hoe vaak zijn labellaars het eens?
# how often do annotators agree?
agreed = []
unagreed = []

for i in range(len(annotated)):
    # als alle labels hetzelfde zijn, dan zijn de labellaars het eens
    # if all labels are the same, annotators agree
    if annotated['Label1'].iloc[i] == annotated['Label2'].iloc[i] == annotated['Label3'].iloc[i]:
        agreed.append(i)
    # anders niet
    # otherwise, they unagree
    else:
        unagreed.append(i)

# print resultaten
# print results
print(len(agreed), len(unagreed))
```

```python
# data in de goede vorm krijgen
# put data in correct format
z = np.array([annotated.Label1, annotated.Label2, annotated.Label3]).T
mask = (z == 4.)
z[mask] = 2.
mask2 = (z == 7.)
z[mask2] = 3.

from statsmodels.stats.inter_rater import fleiss_kappa
from statsmodels.stats.inter_rater import aggregate_raters

z = aggregate_raters(z, n_cat=None)

# waarde van fleiss' kappa berekenen
# calculate fleiss' kappa
fleiss_kappa(z[0])
```
